```python
import cv2
import matplotlib.pyplot as plt

image_path = "Data/Apollo_11_Launch.jpg"
image = cv2.imread(image_path)

#Cropping image
cropped = image[0:900, 300:900] #[start row:end row, start col;end col] heightxwidth
#Changing brightness
bright = cv2.convertScaleAbs(cropped, alpha=1.0, beta=67)
#Rotating image
(h, w) = bright.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(bright, M, (w, h))

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

#Saving modified image
cv2.imwrite("Modified_Apollo_11_Launch_KMQ18.jpg", rotated)
```
